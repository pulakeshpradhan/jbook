# Xarray Aggregating Time Series

This notebook demonstrates how to aggregate time series data using Xarray.

```python
import xarray as xr
import numpy as np
import pandas as pd

# Create sample data
times = pd.date_range("2000-01-01", periods=365, freq="D")
data = xr.DataArray(np.random.rand(365), coords=[times], dims=["time"])

# Aggregate to monthly
monthly = data.resample(time="M").mean()
monthly
```
